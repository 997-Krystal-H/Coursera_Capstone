# Part 1: Pre-processing Data

```python
myurl = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
source = requests.get(myurl).text
soup = BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
table = str(soup.table)
display_html(table,raw=True)
```

```python
df = pd.read_html(table)
df=df[0]
df.head()
```

```python
# Dropping the rows where Borough is 'Not assigned'
df = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
# make sure they are separated by a comma
df = df.groupby(['Postal Code'], sort=False).agg(','.join) 
df.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough. If not, it is still "Neighborhood"
df['Neighbourhood'] = np.where(df['Neighbourhood'] == 'Not assigned',df['Borough'], df['Neighbourhood'])
df
```

```python
# size of dataframe
df.shape
```

# Part 2: Using the csv file to create dataframe

```python
# read file
Geo = pd.read_csv('https://cocl.us/Geospatial_data')
Geo.head()
```

```python
# merge two dataframs
dfNew = pd.merge(df, Geo, on='Postal Code')
```

```python
dfNew
```

# Part 3: Using k-means for Clustering 

```python
# get rows containing the key word "Downtown Toronto"
dfNew = dfNew[dfNew['Borough'].str.contains('Downtown Toronto', regex=False)]
dfNew
```

```python
# visualization

# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[43.651070,-79.347015], zoom_start=10)

# add markers to map
for lat,lng,borough,neighbourhood in zip(dfNew['Latitude'],dfNew['Longitude'],dfNew['Borough'],dfNew['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
# drop 3 catagorical variables in this dataset
dfNew2 = dfNew.drop(['Postal Code','Borough','Neighbourhood'], axis=1)

# normalize the dataset
from sklearn.preprocessing import StandardScaler
X = dfNew2.values[:,1:]
X = np.nan_to_num(X)
cluster_dataset = StandardScaler().fit_transform(X)
cluster_dataset

# apply k-means
num_clusters = 5
k_means = KMeans(init="k-means++", n_clusters=num_clusters, n_init=12)
k_means.fit(cluster_dataset)
labels = k_means.labels_
print(labels)
```

```python
# create a new table
dfNew.insert(0, 'Cluster Labels', labels)
dfNew.head()
```
